```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
np.random.seed(10)
```

```python
randDF = DataFrame(np.random.rand(100).reshape(10,10))
```

```python
randDF[[0,3,5]].mean().round(2)
```

```python
randDF = randDF.round(2)
```

```python
randDF.std().round(2)
```

```python
plt.boxplot(data=randDF, x=randDF, meanline=randDF[0].mean())
plt.show()
```

```python
sns.boxenplot(data=randDF)
```

```python
randDF.median().round(2)
```

```python
randDF.quantile(.75).round(2)
```

```python
randDF.quantile(.25).round(2)
```

```python
sns.barplot(data=randDF)
```
